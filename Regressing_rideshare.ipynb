```python
def impute_price(cols):
    Price = cols[0]
    Hour = cols[1]
    if pd.isnull(Price):
        if Hour >= 0 and Hour < 200:
            return 10
    else:
        return Price
```

```python
ride['price'] = ride[['price','hour']].apply(impute_price,axis=1)
```

```python
ride.dropna(inplace=True)
```

```python
# price null values are handled
ride.columns
```

```python
# hour->price
X = ride[['id', 'timestamp', 'hour', 'day', 'month', 'datetime', 'timezone','source', 'destination', 'cab_type', 'product_id', 'name','distance', 'surge_multiplier', 'latitude', 'longitude', 'temperature']]
```

```python
y = ride['price']

```

```python
X.drop('id',axis=1,inplace=True)
```

```python
X
```

```python
X.drop(['datetime','timezone', 'source','destination','cab_type'],axis=1,inplace=True)
```

```python
X.drop
```

```python
X.drop(['product_id','name'],axis=1,inplace=True)
```

```python
X
```

```python
X.drop('surge_multiplier',axis=1,inplace=True)
```

```python
X.drop('timestamp',axis=1,inplace=True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
print('The coefficients are : ',reg.coef_)
```

```python
print('Intercept : ',reg.intercept_)
```

```python
prediction = reg.predict(X_test)
```

```python
plt.scatter(y_test,prediction)
plt.xlabel('Y_test')
plt.ylabel('predicted values')
```

```python
# difference between them is minimal as they are almost collinear going upwards
```

```python
from sklearn import metrics

print('MAE: ',metrics.mean_absolute_error(y_test,prediction))
print('MSE: ',metrics.mean_squared_error(y_test,prediction))
print('RMSE: ', np.sqrt(metrics.mean_absolute_error(y_test, prediction)))
```

```python
sns.distplot(y_test-prediction,bins=40)
```
